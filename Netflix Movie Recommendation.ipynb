### Genre Feature

```python
df['genres'][0]
```

```python
type(df['genres'][0])
```

```python
# Removes the string from the data
ast.literal_eval(df['genres'][0])
```

```python
# i iterates in the list of genres[0]
# hence i holds each dictionary element of list
# i['name'] is the key of dictionary and prints its genre
for i in ast.literal_eval(df['genres'][0]):
    print(i['name'])
```

```python
# Making function
def genres(data):
    lst= []
    for i in ast.literal_eval(data):
        lst.append(i['name'])
    return lst
genres(df['genres'][0])
```

```python
# Applying this function to all the records of Genre column
df['genres']= df['genres'].apply(genres)
df['genres']
```

### Keyword Feature

```python
df['keywords'][0]
```

```python
# Dictionary elements of Keyword feature is same as Genre
# So same function is applied on it
```

```python
df['keywords']= df['keywords'].apply(genres)
df['keywords']
```

```python
print(df['keywords'][0])
```

### Cast Feature

```python
df['cast'][0]
```

```python
def cast(data):
    counter= 0
    l= []
    for j in ast.literal_eval(data):
        if counter != 5:
            l.append(j['name'])
            counter += 1
    return l
cast(df['cast'][0])
```

```python
# Only top 5 names are included
df['cast']= df['cast'].apply(cast)
df['cast']
```

### Crew Feature

```python
df['crew'][0]
```

```python
def crew(data):
    l= []
    for j in ast.literal_eval(data):
        if j['job'] == 'Director' or j['job'] == 'Producer':
            # To check only for unique names as one person may have both the roles as well
            if j['name'] not in l:
                l.append(j['name'])
    return l
crew(df['crew'][0])
```

```python
df['crew']= df['crew'].apply(crew)
df['crew']
```

## Applying NLP Concepts

```python
# Tokenisation
# Lower Case
# Stemming
# Stopwords Removal
# Encoding to Vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps= PorterStemmer()
ps
```

```python
# Function for performing stemming
def stem(text):
    lst= []
    for i in text.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
new_df['tags'][2]
```

```python
stem(new_df['tags'][2])
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

## BOW Technique

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer(max_features= 5000, stop_words= "english")
cv
```

```python
vector1= cv.fit_transform(new_df['tags'])
vectors= cv.fit_transform(new_df['tags']).toarray()
vector1, vectors
```

```python
#cv.get_stop_words() gives all the stopwords
#cv.get_feature_names_out()
for i in cv.get_feature_names_out():
    print(i)
```

## Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
len(vectors), vectors
```

```python
cos= cosine_similarity(vectors)
cos
```

```python
#Function for recommending top 5 similar movies
def recommend(movie):
    movie_index= new_df[new_df['title'] == movie].index[0]
    distance= cos[movie_index]
    movie_list= sorted(list(enumerate(distance)), reverse= True, key= lambda x : x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
#    for i in movie_list:
#       print(i)
#    print('*******************************************************************************')
#    print()
#    for m in movie_list:
#        print(new_df['title'][m[0]])
#    print('*******************************************************************************')
#    print()
#    for j in movie_list:
#       print(new_df.iloc[j[0]:j[0]+1])
```
